```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../datasets/medical-charges.csv")
```

```python
data["charges"].hist(bins=30)

plt.xlabel("Charges")
plt.ylabel("Count")
```

```python
counts = data["smoker"].value_counts()

plt.bar(counts.index, counts.values)
plt.xlabel("Smoker")
plt.ylabel("Count")
```

```python
plt.scatter(data["bmi"], data["charges"])

plt.xlabel("BMI")
plt.ylabel("Charges")
```

```python
charges = data.groupby("smoker")["charges"].mean()

plt.bar(charges.index, charges.values)

plt.xlabel("Smoker")
plt.ylabel("Mean charges")
```

```python
plt.scatter(data["age"], data["charges"])

plt.xlabel("Age")
plt.ylabel("Charges")
```

```python
charges = data.groupby("region")["charges"].mean()

plt.bar(charges.index, charges.values)

plt.xlabel("Region")
plt.ylabel("Mean charges")
```

```python
charges = data.groupby("sex")["charges"].mean()

plt.bar(charges.index, charges.values)

plt.xlabel("Sex")
plt.ylabel("Mean charges")
```
